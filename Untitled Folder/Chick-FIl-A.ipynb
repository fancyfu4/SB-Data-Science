# What is the healthiest option on a Chick-Fil-A Menu?

Everybody has fast food once in a while, whether its McDonalds, Jack in the Box or even Taco Bell. We all know its not the healthiest choice but it is extremely convenient when you are on the go or just don't have time to cook a full meal. In our family, we would most often choose to get fast food if that day is particularly busy and we are already very close to dinner time. We still want our family to eat healthy so I found an article on the "healthiest" fast foods, among the top three were Chipotle, Taco Bell and Chick-Fil-A. My family doesn't frequent Chipotle or Taco Bell, but we do love Chick-Fil-A! This also makes me wonder just how healthy is Chick-Fil-A foods? 

Through this investigation, I want to answer the following.. 
What is the most unhealthy thing on Chick-Fil-A menu?

First I will import the necessary packages and open up our file. 

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = ('chick-fil-a-nutrition.csv')
df=pd.read_csv(file, sep=',')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

Exploring the data a little bit shows us that there are 290 rows with 12 columns

```python
df[df.duplicated() == True]
```

```python
missing = pd.concat([df.isnull().sum(), 100 * df.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(by='count')
```

The data looks relatively clean as there are no missing values and no duplicated values.

Now to take a look at the data based off of calories in order to determine what is the healthiest option!

```python
cal=df.set_index('Menu')['Calories'].sort_values(ascending=False)
```

```python
print(cal)
```

Taking a further look into this data, I find that the data includes all the catering options as we see the current top 5 highest calorie options include Mac and Cheese trays or other trays. I also notice that the lowest calorie things on the menu include things like pickles and a 5-lb bag of ice. While these are interesting to keep in mind, I want to focus this investigation into healthiest options for a single person, which means we can remove the catering options.

```python
df=df[df['Menu'].str.contains('Tray')==False]
```

```python
cal=df.set_index('Menu')['Calories'].sort_values(ascending=False)
print(cal)
```

```python
df=df[df['Menu'].str.contains('Gallon')==False]
```

```python
cal=df.set_index('Menu')['Calories'].sort_values(ascending=False)
print(cal)
```

```python
df=df[df['Menu'].str.contains('8oz')==False]
```

I continue to remove a couple more items as a Gallon of Chick-Fil-A beverages and 8oz sauce containers would not be what a single person would order.

```python
cal=df.set_index('Menu')['Calories'].sort_values(ascending=False)
print(cal)
```

This leaves us with the top 5 highest calorie items being
1. 30 ct Chick-Fil-A Nuggets
2. Cobb Salad with Chick-Fil-A Chick-n-Strips
3. Spicy Southwest Salad w/ Chick-Fil-A Chick-n-strips
4. 10 ct Chick-Fil-A Chick-n-Minis
5. Cobb Salad w/ Spicy Filet

This surprised me, as I did not expect the salads to be within the highest calorie items. 

I will use this current set of data and look into the sodium content and fat content. Since we are looking into what a single person would order, I am comfortable leaving out the tray, gallon and 8oz sauces that I originally removed when looking at the calorie content. 

```python
sod=df.set_index('Menu')['Sodium (MG)'].sort_values(ascending=False)
print(sod)
```

I was still surprised to see that the salads fill up 4 of the 5 spots on the top 5 for fat content. 

1. 30 ct Chick-fil-A® Nuggets
2. Cobb Salad w/ Spicy Filet
3. Spicy Southwest Salad w/ Spicy Filet
4. Cobb Salad
5. Cobb Salad w/ Nuggets 

```python
fat=df.set_index('Menu')['Fat (G)'].sort_values(ascending=False)
print(fat)
```

I feel less surprised seeing that the salads filled the top 5 spots for highest fat content. I suppose that cobb salads are more calorie dense and have more fat and sodium content than I had originally expected. 

1. Cobb Salad w/ Chick-fil-A Chick-n-Strips
2. Cobb Salad w/ Spicy Filet
3. Cobb Salad w/ Chick-fil-A® Filet
4. Spicy Southwest Salad w/ Chick-fil-A Chick-n-Strips
5. Cobb Salad

```python
carb=df.set_index('Menu')['Carbohydrates (G)'].sort_values(ascending=False)
print(carb)
```

This was not as surprising to see that milkshakes fill 4 out of 5 of the top 5 carb spots. 

1. Strawberry Milkshake
2. 10 ct Chick-fil-A Chick-n-Minis
3. Cookies & Cream Milkshake
4. Chocolate Milkshake
5. Vanilla Milkshake

```python
plt.figure(figsize=(50,10))
sns.boxplot(y='Calories',x='Menu',data=df)
plt.title('Chickfila calorie overview')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(50,10))
sns.boxplot(y='Sodium (MG)',x='Menu',data=df)
plt.title('Chickfila sodium content overview')
plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(50,10))
sns.boxplot(y='Fat (G)',x='Menu',data=df)
plt.title('Chickfila fat content overview')
plt.xticks(rotation=90)

plt.show()
```

Through this investigation we found that variations of the Cobb salad dominated the top spots for unhealthiest item on the Chick-Fil-A menu. 
With the Cobb Salad w/ Chick-fil-A Chick-n-Strips at 910 calories
Cobb Salad w/ Spicy Filet  with 2390 mg sodium
Cobb Salad w/ Chick-fil-A Chick-n-Strips with 63g fat
Cobb Salad w/ Spicy Filet with 62 g fat

And the Milkshakes took over the top spots for carbs. 

With this information, I would say that the most unhealthy meal you could have at Chick-Fil-A would be the Cobb Salad with Chick-n-Strips and a strawberry milkshake. 

This could be considered a good thing that the unhealthiest item is a salad, as salads are usually considered a healthier choice when it comes to fast meals!
