```python
def scatterplots(columns, ncol=None, figsize=(15, 8)):
    if ncol is None:
        ncol = len(columns)
    nrow = int(np.ceil(len(columns) / ncol))
    fig, axes = plt.subplots(nrow, ncol, figsize=figsize, squeeze=False)
    fig.subplots_adjust(wspace=0.5, hspace=0.6)
    for i, col in enumerate(columns):
        ax = axes.flatten()[i]
        ax.scatter(x = col, y = 'Life Expectancy', data=health_data, alpha=0.5)
        ax.set(xlabel=col, ylabel='Life Expectancy')
    nsubplots = nrow * ncol    
    for empty in range(i+1, nsubplots):
        axes.flatten()[empty].set_visible(False)
```

```python
features = [f for f in health_data.columns if f not in ['State', 'County', 'Life Expectancy']]
```

```python
scatterplots(features, ncol=4, figsize=(15, 15))
plt.title('Life Expectancy Scatterplots')
```

Taking a closer look at just the features plotted with the life expectancy, I was also really surprised to see that the % unemployed and % uninsured were so flat. I would have assumed that a higher amount of unemployed or a higher amount of uninsured people would result in lower life expectancy. But perhaps this is a good sign for our healthcare that uninsured and unemployed are not affecting life expectancy!
